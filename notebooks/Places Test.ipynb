```python
# Because header rows are split, some columns are unnamed
df.columns[0:]
```

```python
# fix the split column header name, clean up the table
df.columns = df.columns + ' ' + df.iloc[0,:]
df = df.iloc[1:].reset_index(drop=True)
```
